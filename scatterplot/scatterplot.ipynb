```python
import numpy as np
import csv
```

```python
dataPath = 'data.csv'
```

```python
def get_data(path):
    f = open(path, 'r')
    reader = csv.reader(f, delimiter=',')
    header = next(reader)
    data = []
    for line in reader:
      
        data.append([float(x) for x in line])
        
    return header, np.transpose(data)
```

```python
header, data = get_data(dataPath)
```

```python
len(data[0])
```

```python
import matplotlib.pylab as plt
```

```python

x = data[0] #composition
y = data[1] #band gap
z = data[2] #film thickness
```

```python
m, c = np.polyfit(x, y, 1)
print('m = {0:.2f}, c = {1:.2f} eV'.format(m, c))
```

```python
% pylab inline
fig = figure(figsize=(12,7))
ax = fig.add_subplot(111)


l1 = ax.scatter(x,y,c=z,s=50)
cb = colorbar(l1)
cb.set_label('{0}'.format(header[2]), size=20)
ax.set_xlabel('{0}'.format(header[0]), fontsize = 20)
ax.set_ylabel('{0}'.format(header[1]), fontsize = 20)
ax.tick_params(labelsize=15)
cb.ax.tick_params(labelsize=15)

linRegx = np.linspace(0, 1, 100)
linRegy = m*linRegx + c

ax.plot(linRegx, linRegy, '-', color='red')
ax.set_xlim([0,1])
```
